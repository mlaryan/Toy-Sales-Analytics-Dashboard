```python
def box_histogram(x):
    plt.figure(figsize=(12,5)) 
    
    plt.subplot(1,3,1)
    sns.histplot(data=df ,x=x, kde=True)
    plt.xticks(rotation=45)
    
    plt.subplot(1,3,2)
    sns.boxplot(data=df , y=x)
    plt.tight_layout()
```

```python
box_histogram("Product_Cost"),

box_histogram("Product_Price")
```
